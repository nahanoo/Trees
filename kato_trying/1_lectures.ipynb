```python
#using cache
#@lru_cache(maxsize=10) #add the decoration - simply asking python to use cache memory you can save a lot of time
def fib(n):
    if n==0:
        return 0
    elif n==1 or n==2:
        return n
    else:
        return(fib(n-1)+fib(n-2))
```

```python
fib(6)
```

```python
import time
start=time.time()
for n in range(30):
    print(fib(n))
end=time.time()
print(f"time it took: {end-start} seconds")
```
